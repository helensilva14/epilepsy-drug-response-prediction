```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

LMAX = 16
LMAX2 = LMAX + 4 
LSMIN = sigmoid(-LMAX2)
LSMAX = sigmoid(LMAX2)
def logit(x):
    xc = np.clip(x, LSMIN, LSMAX)
    lg = np.clip(np.log(xc) - np.log(1 - xc), -LMAX, LMAX)
    return lg


# Pega apenas amostras classificadas como sendo da classe 0 (damage)
y_prob_damage = y_prob[y_test==0]

# Pega a probabilide da amostra ser da classe 0 (provavelmente alta probabilidade)
y_prob_damage = y_prob_damage[:,0]
logit_damage = list(map(logit,y_prob_damage))

# Pega apenas amostras classificadas como sendo da classe 1 (regular)
y_prob_regular = y_prob[y_test==1]

# Pega a probabilide da amostra ser da classe 0 (provavelmente baixa probabilidade)
y_prob_regular = y_prob_regular[:,0]
logit_regular = list(map(logit,y_prob_regular))

# for myidx,prob in enumerate(y_prob_damage):
#     line = str(prob) + "\t" + str(logit_damage[myidx]) + "\t" + str(logit2_damage[myidx]) + "\t\n"
#     fp.write(line)
# fp.close()


plt.figure(figsize=(12, 10))
plt.title('CDF Model Trained on Mobilenet')
dist = logit_damage[:]
bins = np.sort(dist)
frequency = np.arange(1,len(bins)+1)/len(bins)
plt.plot(bins, frequency,'r', label="CDF - Blurred")

# dist = logit2_damage[:]
# bins = np.sort(dist)
# frequency = np.arange(1,len(bins)+1)/len(bins)
# plt.plot(bins, frequency,'black', label="CDF - Damage - Mine")


dist = logit_regular[:]
bins = np.sort(dist)
frequency = np.arange(1,len(bins)+1)/len(bins)
plt.plot(bins, frequency, 'b', label="CDF - Not Blurrred")


plt.legend(loc = 'upper left')
plt.xlim([-16.0001, 16.0001])
plt.xticks(np.arange(-16.0001,16.0001,1),rotation=60)

plt.ylim([0, 1.01])
plt.yticks(np.arange(0, 1, step=0.05))
plt.ylabel('CDF')
plt.xlabel('Logit Damage')
plt.grid(True)

dst = "./graphics/"
print("\n <------ Saving CDF Curve at %s ------>" % (dst))
if os.path.isdir(dst) == False:
    os.makedirs(dst)
dst = dst + "/LOGIT_Mobilenet_from_scratch.png"
plt.savefig(dst)
```
